```bash
#project="Murthy"

base_dir=/mnt/basespace/Projects/Murthy


# Murthy mut projecrt

for sample in $(ls $base_dir/Samples/ | grep "^IS-032")
do
    src=$base_dir/Samples/$sample/Files
    dest="../data/$(echo $sample | cut -d'-' -f3)"
    mkdir -p $dest
    echo "Downloading into $dest ..."
    cp $src/*.fastq.gz $dest
done
```

```bash
# Combine reads from 4 lanes into 1 file

base_dir="../data"
for sample in $(ls $base_dir)
do
    for f in $(ls $base_dir/$sample | grep R1)
    do
        cat $base_dir/$sample/$f >> $base_dir/$sample/R1.fastq.gz
    done
    
    for f in $(ls $base_dir/$sample | grep R2)
    do
        cat $base_dir/$sample/$f >> $base_dir/$sample/R2.fastq.gz
    done
    rm -rf $base_dir/$sample/skm*.fastq.gz
    ls -lah $base_dir/$sample
done
```
